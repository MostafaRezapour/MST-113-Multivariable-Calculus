```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # permet d'utiliser l'attribut projection="3d"

x = np.linspace(-3, 3, 200)
y = x.copy()
X, Y = np.meshgrid(x, y)

Z = 100*(Y-X**2)**2+(1-X)**2

fig=plt.figure(figsize=(15,10))
ax = plt.axes(projection="3d")

ax.contour(X, Y, Z, zdir="z", offset=-1)
ax.plot_surface(X, Y, Z, alpha=.8, cmap=cm.coolwarm)


plt.savefig("surface3D_sinFoisSin_python_matplotlib.svg", format="svg")
ax = Axes3D(fig)

def init():
    # Plot the surface.
    ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=20, antialiased=False)
    return fig,

def animate(i):
    # azimuth angle : 0 deg to 360 deg
    ax.view_init(elev=20, azim=i*4)
    return fig,

# Animate
ani = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=90, interval=150, blit=True)
HTML(ani.to_html5_video())
```
